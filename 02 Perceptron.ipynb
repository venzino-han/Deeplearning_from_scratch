## 2.3 퍼셉트론 구현하기

```python
def AND(x1, x2):
    w1, w2, theta = 0.5,0.5,0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(0,0)
```

```python
def AND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = - 0.7
    tmp = np.sum(x*w)+b
    if tmp > 0 :
        return 1
    else :
        return 0
```

```python
AND(0,1)
AND(0.8,0.8)
```

#### W 는 각 입력신호가 결과에 영향을 주는 영향력을 조절,
#### b 는 뉴런이 얼마나 쉽게 활성화(결과로 1을 출력) 하느냐를 조정하는 변수

```python
#AND 결과를 역으로 출력

def NAND(x1,x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    
    tmp = np.sum(w*x) + b
    if tmp <=0:
        return 0
    else:
        return 1
```

```python
NAND(1,1)
NAND(0,1)
NAND(0.8,0.8)
```

```python
#OR : 둘중 하나만이라도 1이면 1 출력 
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    
    tmp = np.sum(w*x) +b
    
    if tmp <= 0:
        return 0 
    else:
        return 1
```

```python
OR(1,0.8)
OR(0,0.7)
OR(0,0.2)
```

### 다층 퍼셉트론의 등장

```python
plt.imshow(imread('004.jpg'))
plt.show()
```

```python
#XOR 게이트 구현
def XOR(x1,x2):
    s1 = NAND(x1,x2) #둘다 1일 때만 0 
    s2 = OR(x1,x2) #둘중 하나만 1이면 1
    y = AND(s1,s2) #둘다 1일 경우만 1
    
    return y
```

```python
XOR(1,0) # --> 1
XOR(1,1)
XOR(0,0)
XOR(0.8,0.1) # --> 1
```
